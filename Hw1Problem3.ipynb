Problem 3

What does an average MNIST digit look like? For each i=0,1,...,9, compute the average of digit i and display it.

# MNIST digits data set

## Loading the MNIST digits data set

```
from keras.datasets import mnist
import numpy as np

(train_images_org, train_labels), (test_images_org, test_labels) = mnist.load_data()
train_images=train_images_org.reshape((60000,28*28))
train_images=train_images.astype('float32')/255
```

```
import matplotlib.pyplot as plt
```

```
zero=[]
one=[]
two=[]
three=[]
four=[]
five=[]
six=[]
seven=[]
eight=[]
nine=[]
```

```
for i in range(6000):
  if( train_labels[i]==0):
    zero.append(train_images[i])
  if( train_labels[i]==1):
    one.append(train_images[i])
  if( train_labels[i]==2):
    two.append(train_images[i])
  if( train_labels[i]==3):
    three.append(train_images[i])
  if( train_labels[i]==4):
    four.append(train_images[i])
  if( train_labels[i]==5):
    five.append(train_images[i])
  if( train_labels[i]==6):
    six.append(train_images[i])
  if( train_labels[i]==7):
    seven.append(train_images[i])
  if( train_labels[i]==8):
    eight.append(train_images[i])
  if( train_labels[i]==9):
    nine.append(train_images[i]) 
```

```
avg=np.average(zero, axis=0)
avg=avg.reshape((28,28))
plt.imshow(avg)
plt.show()

avg=np.average(one, axis=0)
avg=avg.reshape((28,28))
plt.imshow(avg)
plt.show()

avg=np.average(two, axis=0)
avg=avg.reshape((28,28))
plt.imshow(avg)
plt.show()

avg=np.average(three, axis=0)
avg=avg.reshape((28,28))
plt.imshow(avg)
plt.show()

avg=np.average(four, axis=0)
avg=avg.reshape((28,28))
plt.imshow(avg)
plt.show()

avg=np.average(five, axis=0)
avg=avg.reshape((28,28))
plt.imshow(avg)
plt.show()

avg=np.average(six, axis=0)
avg=avg.reshape((28,28))
plt.imshow(avg)
plt.show()

avg=np.average(seven, axis=0)
avg=avg.reshape((28,28))
plt.imshow(avg)
plt.show()

avg=np.average(eight, axis=0)
avg=avg.reshape((28,28))
plt.imshow(avg)
plt.show()

avg=np.average(nine, axis=0)
avg=avg.reshape((28,28))
plt.imshow(avg)
plt.show()
```
